```python
import ttkbootstrap as ttk
from ttkbootstrap.tableview import Tableview
from ttkbootstrap.constants import *

import tkinter as tk
import sqlite3
from tkinter import messagebox

conn = sqlite3.connect('mycontact.db')
c = conn.cursor()

c.execute("""CREATE TABLE IF NOT EXISTS contacts
                (id INTEGER PRIMARY KEY AUTOINCREMENT,
                 name TEXT,
                 phone TEXT,
                 address TEXT
                 )
            """)
conn.commit()
conn.close()

wf = ttk.Window()
wf.title('contacts app')
wf.geometry('500x500')
colors = wf.style.colors


def view_data():
     dt.delete_rows()
     conn = sqlite3.connect('mycontact.db')
     c = conn.cursor()
     ouput = c.execute("SELECT * FROM contacts")
     for i in ouput.fetchall():
          # print(i)
          dt.insert_row('end', [i[0],i[1], i[2],i[3]])
     conn.commit()
     conn.close()

def create_data():
     a = en.get()
     b =ph.get()
     d =ad.get()
     conn = sqlite3.connect('mycontact.db')
     c = conn.cursor()
     c.execute("INSERT INTO contacts (name, phone, address) VALUES (?,?,?)",(a,b,d))
     conn.commit()
     conn.close()
     messagebox.showinfo('success', 'Data Save')
     view_data()


def delete_data():
     conn = sqlite3.connect('mycontact.db')
     c = conn.cursor()
     selected_rows = dt.get_rows(selected=True)
     print(selected_rows)
     for row_obj in selected_rows:
           row_id = row_obj.values[0]
           c.execute("DELETE FROM contacts WHERE id=?", (row_id,))
     conn.commit()
     conn.close()
     messagebox.showerror('success', 'Data Delete')
     view_data()

def update_data():
     a = en.get()
     b =ph.get()
     d =ad.get()
     selected_rows = dt.get_rows(selected=True)
     conn = sqlite3.connect('mycontact.db')
     
     c = conn.cursor()
     for row_obj in selected_rows:
          row_id = row_obj.values[0]
          c.execute("UPDATE contacts SET name=?, phone=?, address=? WHERE id= ?",(a,b,d,row_id))
     conn.commit()
     conn.close()
     messagebox.showerror('update', 'Data Updated')
     view_data()

name = ttk.Label(wf, bootstyle="inverse-success", text='Name')
phone = ttk.Label(wf, bootstyle="inverse-success", text='Phone ')
address = ttk.Label(wf, bootstyle="inverse-success", text = 'Address')

en = ttk.Entry(wf, bootstyle='success')
ph = ttk.Entry(wf, bootstyle='success')
ad = ttk.Entry(wf, bootstyle='success')

cbtn = ttk.Button(text='Create', bootstyle='success', command=lambda:create_data())
ubtn = ttk.Button(text='Update', bootstyle='info', command=lambda:update_data())
dbtn = ttk.Button(text='Delete', bootstyle='danger', command=lambda:delete_data())



name.pack()
en.pack()
phone.pack()
ph.pack()
address.pack()
ad.pack()
cbtn.pack()
ubtn.pack()
dbtn.pack()


rowdata = []

dt = Tableview(
    master=wf,
    coldata=['ID','Name', 'Phone', 'Address'],
    rowdata=rowdata,
    paginated=True,
    searchable=True,
    bootstyle=PRIMARY,
    stripecolor=(colors.light, None),
)
dt.pack(fill=BOTH, expand=YES, padx=10, pady=10)



view_data()
wf.mainloop()
```
