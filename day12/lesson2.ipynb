```python
# !pip install ttkbootstrap
from tkinter import *
import tkinter as tk
import ttkbootstrap as ttk
from ttkbootstrap.constants import *


cal = tk.Tk()
cal.title('my Calculator App')
cal.geometry('500x400')

# 34+20
abc = "" 

def press_display(num):
    global abc
    abc += str(num)
    result.set(abc)

def answer():
    global abc
    x = eval(abc)  #
    abc = ""
    result.set(x)    

def delete():
    global abc
    abc = ""
    result.set("")
    
def rev():
    global abc  #323
    x = abc[:-1]
    abc = x
    result.set(abc)
    
    
    
    
    

no1 = ttk.Button(cal, text='1',bootstyle="danger",  command=lambda:press_display(1))
no2 = ttk.Button(cal, text='2',  command=lambda:press_display(2))
no3 = ttk.Button(cal, text='3', command=lambda:press_display(3))
no4 = ttk.Button(cal, text='4', command=lambda:press_display(4))

result = StringVar()
display = ttk.Entry(cal, textvariable=result, bootstyle="danger")
display.grid(row=0, column=0, columnspan=4)

no5 = ttk.Button(cal, text='5')
no6 = ttk.Button(cal, text='6')
no7 = ttk.Button(cal, text='7')
no8 = ttk.Button(cal, text='8')

# Row 3: 9 0 ans del
no9  = ttk.Button(cal, text='9',)
no0  = ttk.Button(cal, text='0',)
ans  = ttk.Button(cal, text='ans', command=lambda:answer())
dele = ttk.Button(cal, text='del', command=lambda:rev() )

add = ttk.Button(cal, text='+' , command=lambda: press_display('+'))
sub = ttk.Button(cal, text='-' ,command=lambda: press_display('-'))
mul = ttk.Button(cal, text='*' , command=lambda: press_display('*'))
div = ttk.Button(cal, text='%' )

no1.grid(row=1, column=1,padx=4, pady=4)
no2.grid(row=1, column=2,padx=4, pady=4)
no3.grid(row=1, column=3,padx=4, pady=4)
no4.grid(row=1, column=4,padx=4, pady=4)

no5.grid(row=2, column=1,padx=4, pady=4) 
no6.grid(row=2, column=2,padx=4, pady=4)
no7.grid(row=2, column=3,padx=4, pady=4) 
no8.grid(row=2, column=4,padx=4, pady=4) 

no9.grid(row=3, column=1,padx=4, pady=4) 
no0.grid(row=3, column=2,padx=4, pady=4) 
ans.grid(row=3, column=3, padx=4, pady=4) 
dele.grid(row=3, column=4, padx=4,pady=4)

add.grid (row=4, column=1, padx=4, pady=4)
sub.grid(row = 4, column=2, padx = 4, pady=4)
mul.grid (row=4, column=3, padx=4, pady=4)
div.grid(row =4, column=4, padx=4, pady=4)



cal.mainloop()
```
