```python
import matplotlib.pyplot as plt
a= [5,10,15,20,25]
b = [10,20,23,2,1]
x = [25,20,10,3,4]
y = [3,67,23,12,5]
# subplot(row, col, index) 3x3
plt.subplot(2,3,1)
plt.plot(a,b)
plt.title('Num 1')

plt.subplot(2,3,2)
plt.plot(x,y)
plt.title('plot 2')

plt.subplot(2,3,3)
plt.plot(a,x)
plt.title('plot 3')

plt.subplot(2,3,4)
plt.plot(b,y)
plt.title('plot 4',loc='left')

plt.suptitle('This is Sub Title')
plt.show()
```
